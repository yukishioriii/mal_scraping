```python
import re
from tqdm import tqdm
```

```python
anime = open('mal_scraper/anime.txt').read()
```

```python
anime = re.split("<end>\n", anime)
```

```python
anime_dict = {}
```

```python
for a in anime:
    a = a.split('\n')
    url = a[0]
    
    content = " ".join(a[1:])
    anime_dict[url] = content
```

```python
len(anime_dict)
```

```python
recs = open('mal_scraper/recommendations_full.txt').read()
recs = re.split("<end>\n\n", recs)
```

```python
string = []
for rec in recs:
    try:
        rec = rec.split('\n')
        a1, a2 = rec[0].split(';')
        a1_name = " ".join(a1.split('/')[-1].split("_"))
        a2_name = " ".join(a2.split('/')[-1].split("_"))
        content = " ".join(rec[1:])
        string.append( f"<a1n> {a1_name} <a1> {anime_dict[a1]} <a2n> {a2_name} <a2> {anime_dict[a2]} <content> {content} <end>\n")
    except Exception as e:
        print(rec, e)

with open('rec_anime_data.txt', 'w+') as f:
    f.write("".join(string))
```
